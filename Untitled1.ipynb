```python
normal = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(normal)
```

```python
def get_image(image, cmap= None):
    plt.axis('off')
    plt.imshow(image, cmap=cmap)
    plt.show()
```

```python
#get_image(img) voltar ao rgb
get_image(normal)
```

```python
#colorindo do bgr para hsv
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
get_image(hsv)
```

```python
#escala de cinza
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
get_image(gray, cmap='gray')
```

```python
#borrando imagem - blurred
width, height = normal.shape[:2]
if width > 500:
    k = (50,50)
elif width > 200 and width <=500:
    k = (10,10)

#uso do kernel
blur= cv2.blur(normal, k)
get_image(blur)
```

```python
#preto e branco binario
v, binary = cv2.threshold(gray, 100, 188, cv2.THRESH_BINARY)
get_image(binary, cmap='gray')
```

```python
#invertendo cores
invert = cv2.bitwise_not(binary)
get_image(invert, cmap='gray')
```
